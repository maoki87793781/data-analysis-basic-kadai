```python
import matplotlib.pyplot as plt
import numpy as np

# プロット
#fig, ax = plt.subplots()
fig, axes = plt.subplots(2,3,tight_layout=True)

#折れ線グラフ
y0 = [1, 2, -5, 2]
axes[0,0].plot(y0)

#sin関数
# データを作成
# 0から10までの幅で配列を等間隔に100個作成する
x1 = np.linspace(0, 10, 30)
y1 = 2 + 2 * np.sin(2 * x1)
axes[0,1].plot(x1, y1, linewidth=2.0)

#ヒストグラム
# 平均が15, 標準偏差5の標準正規分布から2000個の乱数の生成する
x2 = np.random.normal(15, 5, 2000)
axes[0,2].set(xlim=(-10,40))
axes[0,2].hist(x2, bins=30)

#散布図
# データを生成する
np.random.seed(3)
x3 = 4 + np.random.normal(0, 2, 60)
y3 = 4 + np.random.normal(0, 2, len(x3))

axes[1,0].scatter(x3, y3)

# プロットする
plt.show()
```
