```python
!pip install japanize_matplotlib

import matplotlib.pyplot as plt
import japanize_matplotlib
import pandas as pd

df = pd.read_csv('sample_pandas_6.csv')
category_df = pd.read_csv('category.csv')

df = pd.merge(df, category_df[['商品番号', 'カテゴリー']], how='inner', on='商品番号')
df
```

```python
# カテゴリーごとの商品数を集計
category_counts = df['カテゴリー'].value_counts()

# 棒グラフの作成
#plt.figure(figsize=(10, 6))  # グラフのサイズを指定
category_counts.plot(kind='bar')
plt.title('カテゴリーごとの商品数')  # グラフのタイトル
plt.xlabel('カテゴリー')  # x軸のラベル
plt.ylabel('商品数')  # y軸のラベル
plt.show()
```

```python
# 商品番号ごとに注文数の基本統計量を計算
order_sum = df.groupby('商品番号')['注文数'].sum()
order_sum
```

```python
# 商品番号ごとに注文数の基本統計量を計算
order_stats = df.groupby('商品番号')['注文数'].describe()
order_stats
```
